```python
import requests
import urllib
from bs4 import BeautifulSoup

links=[]
title=[]
review=[]
stocks=[]
price=[]
stars=[]

def Books(soop):
    t=soop.find_all('p',class_='book_name')
    pric=soop.find_all('p',class_='price green')
    stock=soop.find_all('p',class_='stock')
    reviews=soop.find_all('p',class_='review green')
    books = soop.find_all('div',{'class','book_container'})
    for book in books:
        stars.append(5 - len(book.find_all('span',{'class','not_filled'})))
    
    for a in t:
        title.append(a.text)
        links.append(a.find('a').get("href"))
        
    for p in pric:
        price.append(p.text)
        
    for s in stock:
        stocks.append(s.text)
        
    for r in reviews:
        review.append(r.text)
        
    
        
        
urls = ['https://arifpucit.github.io/bss2/index.html', 
        'https://arifpucit.github.io/bss2/SP.html', 
        'https://arifpucit.github.io/bss2/CA.html']        
print(len(title),len(links),len(price),len(review),len(stocks),len(stars))
for url in urls:
    resp=requests.get(url)
    soop=BeautifulSoup(resp.text,'lxml')
    Books(soop)

print(len(title),len(links),len(price),len(review),len(stocks),len(stars))
data = {'Title/Author':title, 'Price':price, 'Availability':stocks, 
        'Reviews':review, 'Links':links, 'Stars':stars}
df = pd.DataFrame(data, columns=['Title/Author', 'Price', 'Availability', 'Reviews', 'Links', 'Stars'])
df.to_csv('books3.csv', index=False)
df = pd.read_csv('books3.csv')
df
```
